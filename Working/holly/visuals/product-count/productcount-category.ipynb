```python
%matplotlib notebook
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import plotly
# import plotly.plotly as py
# import plotly.graph_objs as go
```

```python
# Load in csv
catcount_df = pd.read_csv("../data/catcount.csv")
catcount_df.head()
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(catcount_df))
tick_locations = [value for value in x_axis]
```

```python
plt.figure(figsize=(15,8))
plt.bar(x_axis, catcount_df["Count"], color='red', alpha=0.5, align="center")
plt.xticks(tick_locations, catcount_df["Category"], rotation="45", horizontalalignment="right")
plt.title("Product Count by Category", size=26)
plt.xlabel("Category", size=14)
plt.ylabel("Number of Advertised Products", size=14)
plt.tight_layout()
plt.savefig("../../place-plots-here/productcount-category.png")
plt.savefig("../visuals/productcount-category.png")
```
